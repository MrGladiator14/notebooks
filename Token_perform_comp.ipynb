```python
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression

texts = [
    'book a flight',
    'cancle my ticket',
    'weather today',
    'book hotel',
    'cancel reservation',
    'weather tomorrow'
]

labels = ['booking', 'cancer', 'weather', 'booking', 'cancel', 'weather']

def tokenizer_simple(text):
  return text.split()

def tokenizer_custom(text):
  import re
  text = text.lower()
  text = re.sub(r'[^a-z ]', '', text)
  return text.split()
```

```python
def run_pipeline(tokenizer):
  vec = TfidfVectorizer(tokenizer=tokenizer)
  X = vec.fit_transform(texts)
  model = LogisticRegression()
  model.fit(X, labels)

  pred = model.predict(X)
  print(classification_report(labels, pred))

run_pipeline(tokenizer_simple)
run_pipeline(tokenizer_custom)
```
